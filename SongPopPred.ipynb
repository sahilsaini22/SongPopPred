```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import training data
training_data = pd.read_csv('train.csv')
```

```python
training_data
```

```python
##Standardise data
from sklearn.preprocessing import StandardScaler
training_data[['loudness', 'tempo' , 'duration_ms']] = StandardScaler().fit_transform(training_data[['loudness', 'tempo' , 'duration_ms']])


training_data['popularity'] = round((training_data['popularity'] / 10))

training_data
```

```python
#Split into training and validation data
from sklearn.model_selection import train_test_split
train_features, valid_features, train_labels, valid_labels=train_test_split(
    training_data.drop(labels=['year','artists', 'id', 'name', 'release_date', 'popularity'], axis=1),
    training_data['popularity'],
    test_size=0.2,
    random_state=41)
```

```python
#Create Model
from sklearn.svm import SVC
model = SVC()
```

```python
#Fit model to training data
model.fit(train_features, train_labels)
```

```python
#Set Parameteres
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 
```

```python
#Find out the best values for parameters
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
grid.fit(valid_features,valid_labels)
```

```python
#Set the best parameters
grid.best_params_
```

```python
test_data = pd.read_csv('test.csv')

test_labels = round(test_data['popularity'] / 10)

test_data = test_data.drop(['year','artists', 'id', 'name', 'release_date', 'popularity'], axis=1)
```

```python
##Standardise data
from sklearn.preprocessing import StandardScaler
test_data[['loudness', 'tempo' , 'duration_ms']] = StandardScaler().fit_transform(test_data[['loudness', 'tempo' , 'duration_ms']])


test_data
```

```python
#Test the model
grid_predictions =grid.predict(test_data)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(test_labels,grid_predictions))
```
